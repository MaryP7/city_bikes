```python
# Secure all items in the response that correspond to TAVG

avg_temps = [item for item in results if item['datatype']=='TAVG']
```

```python
len(avg_temps)
```

```python
# Get only the date field from all average temperature readings

dates = [item['date'] for item in avg_temps]
```

```python
# Get the temperature from all average temperature readings

temps = [item['value'] for item in avg_temps]
```

```python
df_temp=pd.DataFrame()
```

```python
df_temp['date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in dates]
```

```python
df_temp['avgTemp'] = [float(v)/10.0 for v in temps]
```

```python
files=os.listdir("data/unzipped")
```

```python
csv_files=[f for f in files if f.endswith(".csv")]
csv_files
```

```python
df=pd.DataFrame()

for f in csv_files:
    loaded=pd.read_csv(f"data/unzipped/{f}")
    
    df=pd.concat([df,loaded])
    
df.shape
```
