```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/data_subset.csv", index_col=0)
```

```python
df["trip_duration"]=pd.to_datetime(df["ended_at"])-pd.to_datetime(df["started_at"])
```

```python
df["minutes"]=(df["trip_duration"].dt.days*24*60)+(df["trip_duration"].dt.seconds/60)
```

```python
df
```

Created a value and group by start and end station

```python
df=df.sample(5000,replace=True)
df["value"]=1
```

```python
df_group = df.groupby(['start_station_name', 'end_station_name',"start_lat","start_lng","end_lat","end_lng"])["value"].count().reset_index()
df_group
```

```python
df_group.rename(columns = {'from_station_name':'start_station_name','to_station_name' : 'end_station_name',
'value': 'trips'}, inplace = True)
```

Create KeplerGl instance

```python
from keplergl import KeplerGl
m = KeplerGl()
```

```python
m.add_data(data=df_group,name="data_1")
```

```python
m
```

```python
config = m.config
```

```python
m.save_to_html(file_name='Divvy Bike Trips Aggregated.html', read_only = False, config = config)
```

```python
import json
with open("config.json", "w") as outfile:
    json.dump(config, outfile)
```
